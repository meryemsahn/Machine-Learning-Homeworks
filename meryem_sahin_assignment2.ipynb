## DATA EXPLORATION & PREPARATION (15 points)

```python
direction_encode = {'right': 0, 'left': 1, 'up': 2, 'straight': 3}

eyewear_encode = {'open': 0, "sunglasses":1}
```

```python
import glob
from PIL import Image
filelist_train = glob.glob('TrainingSet/*.jpg')
filelist_test=glob.glob('TestSet/*.jpg')               #opening files
```

```python
X_train=[]
for fname in filelist_train:
    X_train_dim=np.array(Image.open(fname).convert('L')) 
    X_train.append(X_train_dim.flatten())                #for train take images name and put it in an array with flatten version

split_for_direction_train= [fname.rsplit('_')[1] for fname in filelist_train]
y_train_direction=[]                             #for train split image name and take direction part on a list with encoded version
for i in split_for_direction_train:
    y_train_direction.append(direction_encode[i])

X_test=[]
for fname in filelist_test:
    X_test_dim=np.array(Image.open(fname).convert('L'))    #for test take images name and put it in an array with flatten version
    X_test.append(X_test_dim.flatten())

split_for_direction_test= [fname.rsplit('_')[1] for fname in filelist_test]
y_test_direction=[]                           #for test split image name and take direction part on a list with encoded version
for i in split_for_direction_test:
    y_test_direction.append(direction_encode[i])

split_for_eyewear_train= [fname.rsplit('_')[3] for fname in filelist_train] 
split_for_eyewear_train=[i.rsplit('.jpg')[0] for i in split_for_eyewear_train]

y_train_eyewear=[]
for i in split_for_eyewear_train:       #for train split image name and take eyewear part on a list with encoded version
    y_train_eyewear.append(eyewear_encode[i])

split_for_eyewear_test= [fname.rsplit('_')[3] for fname in filelist_test] 
split_for_eyewear_test=[i.rsplit('.jpg')[0] for i in split_for_eyewear_test]

y_test_eyewear=[]
for i in split_for_eyewear_test:      #for test split image name and take eyewear part on a list with encoded version
    y_test_eyewear.append(eyewear_encode[i])
```
